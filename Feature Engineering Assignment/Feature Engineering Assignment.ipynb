```python
data = pd.read_csv('https://raw.githubusercontent.com/Jayesh-Chavan/Data-Visualization-/main/Pandas%20Data%20Visualization/Attrition1.csv')
df = data.copy()
df.head()
```

```python
cat_col = df[df.columns[df.dtypes == 'object']]
cat_col.head()
```

```python
num_col = df[df.columns[df.dtypes != 'object']]
num_col.head()
```

```python
num_col['Age'].describe()
```

```python
#dfmi.loc[:, ('one', 'second')] = value
#becomes
#dfmi.loc.__setitem__((slice(None), ('one', 'second')), value)
```

```python
# Min-Max Scalar for Age
num_col.loc[0:,'Age'] = (num_col['Age'] - num_col['Age'].min())/ (num_col['Age'].max() - num_col['Age'].min())
```

```python
num_col['DailyRate'].describe()
```

```python
# Standardization of Daily Rate
num_col.loc[0:,'DailyRate'] = (num_col['DailyRate'] - num_col['DailyRate'].mean()) / num_col['DailyRate'].std() 
```

```python
num_col.head()
```

```python
num_col['DistanceFromHome'].describe()
```

```python
num_col['HourlyRate'].describe()
```

```python
#Standardization of Hourly Rate
num_col.loc[0:,'HourlyRate'] = (num_col['HourlyRate'] - num_col['HourlyRate'].mean()) / num_col['HourlyRate'].std()
```

```python
num_col['JobSatisfaction'].describe()
```

```python
num_col['MonthlyIncome'].describe()
```

```python
#Standardization of Monthly Income
num_col.loc[0:,'MonthlyIncome'] = (num_col['MonthlyIncome'] - num_col['MonthlyIncome'].mean()) / num_col['MonthlyIncome'].std()
```
